```python
def rf_incov_ratio(IS):
    EBIT = IS[0]['ebitda']-IS[0]['depreciationAndAmortization']
    inexp = IS[0]['interestExpense']
    incov_ratio = EBIT/inexp
    start = date(2020,8,12)
    tres = web.DataReader(['TB1YR'],'fred',start)
    risk = float(tres.iloc[-1])
    rf = risk/100
    print(rf, incov_ratio)
```
